# Resultados pruebas ANOVA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Los siguientes datasets han sido generados tomando un n√∫mero de samples *n_samples=100*.

```python
df_p2 = pd.read_csv('anova_data_100k.csv')
```

El siguiente ejemplo ha sido generado tomando los valores: $\sigma = 1$ y utilizando la norma de $L_2$.

```python
plt.scatter(df_p2.x, df_p2.y);
```

```python
print('Mean: ', np.mean(df_p2.y))
print('Var: ', np.var(df_p2.y))
```

El siguiente ejemplo ha sido generado utilizando la norma de $L_1$. A la vista de que el $p-valor$ siempre era nulo se ha escogido una $\sigma$ superior, en este caso $\sigma=50$.

```python
plt.scatter(df_p1.x, df_p1.y);
```

```python
print('Mean: ', np.mean(df_p1.y))
print('Var: ', np.var(df_p1.y))
```

En el siguiente gr√°fico puede observarse que la norma 1 es mejor en t√©rminos de convergencia hacia el $p-valor$ que la norma 2.

```python
plt.scatter(df_p2.x, df_p2.y - np.mean(df_p2.y), alpha=0.4)
plt.scatter(df_p2.x, df_p1.y - np.mean(df_p1.y), alpha=0.4);
```

```python
df_p1_10 = pd.read_csv('csv/anova_50k_p1_sigma10.csv')
df_p2_10 = pd.read_csv('csv/anova_50k_p2_sigma10.csv')
```

Parece que la separaci√≥n entre los $p-valores$ para diferentes normas depende del n√∫mero de trayectorias que le damos a cada grupo. En este caso probamos con 10 para cada uno, y comprobamos que para ambas el resultado es mucho m√°s cercano que en el caso anterior.

```python
plt.scatter(df_p2_10.x, df_p2_10.y, alpha=0.4)
plt.scatter(df_p1_10.x, df_p1_10.y, alpha=0.4);
```
